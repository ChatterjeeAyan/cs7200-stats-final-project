```python
import openpyxl

wb = openpyxl.load_workbook('Suppementary_Combined.xlsx')
```

# Hydrophobicity Regression

```python
hydrophobicity_sheet = wb.get_sheet_names()[4]
hydrophobicity_worksheet = wb.get_sheet_by_name(hydrophobicity_sheet)
```

```python
for row in range(2, hydrophobicity_worksheet.max_row+1): 
    column = "F"
    cell_name = "{}{}".format(column, row)
    from_type = hydrophobicity_worksheet[cell_name].value
    column = "G"
    cell_name = "{}{}".format(column, row)
    to_type = hydrophobicity_worksheet[cell_name].value
    if from_type == 'Hydrophobic' and to_type == 'Hydrophilic':
        column = "I"
        cell_name = "{}{}".format(column, row)
        hydrophobicity_worksheet[cell_name].value = -1
    if from_type == 'Hydrophilic' and to_type == 'Hydrophobic':
        column = "I"
        cell_name = "{}{}".format(column, row)
        hydrophobicity_worksheet[cell_name].value = 1
    if from_type == 'Hydrophobic' and to_type == 'Neutral':
        column = "J"
        cell_name = "{}{}".format(column, row)
        hydrophobicity_worksheet[cell_name].value = -1
    if from_type == 'Neutral' and to_type == 'Hydrophobic':
        column = "J"
        cell_name = "{}{}".format(column, row)
        hydrophobicity_worksheet[cell_name].value = 1
    if from_type == 'Hydrophilic' and to_type == 'Neutral':
        column = "K"
        cell_name = "{}{}".format(column, row)
        hydrophobicity_worksheet[cell_name].value = -1
    if from_type == 'Neutral' and to_type == 'Hydrophilic':
        column = "K"
        cell_name = "{}{}".format(column, row)
        hydrophobicity_worksheet[cell_name].value = 1 
```

```python
wb.save('Suppementary_Combined.xlsx')
```
